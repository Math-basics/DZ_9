```python
# Загружаем библиотеки:
# для работы с математикой
import numpy as np
# для построения графиков
import matplotlib.pyplot as plt 
# для раболты со статистикой
from scipy import stats
```

```python
# Создаём случайную величину из нормального распределения
norm_rv = stats.norm(loc = 0, scale = 1)

# Зададим несколько значений n (n - размер выборки)
# и словарь результатов
result = {1: [], 2: [], 3: [], 5: [], 10: [], 20: [], 30: [], 50: [], 100: []}

# Перебираем последовательно все значения n
for n in result.keys():
    # Создаём 1000 выборок размера n, рассчитываем и записываем среднее выборочное
    for i in range(1000):
        result[n].append(norm_rv.rvs(n).mean())
```

```python
# График распределения средних выборочных

# Создаём поле для графиков
fig, axes = plt.subplots(nrows = len(result), # Количество графиков соответствует n
                         ncols = 1, # В строке 1 график
                         figsize=(7, 20), # Размер общего поля
                         sharex = 'all' # Делаем ось X общей для всех графиков
                        ) 

# Счётчик позиций графиков
plot_n = 0
# Строим графики для каждого n
for n in result.keys():
    axes[plot_n].hist(result[n],
                     bins = 20,
                     density = True
                     )
    axes[plot_n].set_xlim(-3, 3)
    axes[plot_n].set_title(f'n = {n}', fontsize = 'x-large')
    # Нормальное распределение для сравнения
    axes[plot_n].plot(np.linspace(-3, 3, 1000), stats.norm.pdf(np.linspace(-3, 3, 1000)))
    plot_n += 1
    
fig.suptitle('\n\n\nГистограммы распределения средних выборочных в зависимости от размера выборки n', fontsize = 'xx-large')  
axes[plot_n - 1].set_xlabel('Диапазоны значений', fontsize = 'xx-large')  

plt.show()

print('Распределение средних выборочных похоже на нормальное распределение только при единичном размере выборки n = 1.')
print('С увеличением размера выборки n распределение средних выборочных сужается и приближается к математическому ожиданию для генеральной совокупности.')
```
